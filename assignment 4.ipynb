```python
list = [12 , 1 , 8 , -1 , 7 , 8 , 1 , -8 , 1  , 9]
len(list)

```

```python
#minimum value
minval = list[0]
for i in list:
    if i < minval:
        minval = i
print(minval)
```

```python
# max value
maxval = list[0]
for i in list:
    if i > maxval:
        maxval = i
print(maxval)
```

```python
# Average
#sum function used to add all values in the list in python
sum_list = sum(list)
val = len(list)
ans = sum_list / val
print(ans)
```

```python
# value repeated in number of times
val = list[2]
a = 0
for i in list:
    if val == i:
        a += 1
print(a)
        
```
