```python
cities = ['karachi','islamabad','quetta','kashmir','gilgit','lahore','naran','kaghan','nathia','faislabad']
city_k = []
city_nk = []
for city in cities:
    for char in city:
        if char == 'k':
            city_k.append(city)
for cite in cities:
    for citi in city_k:
        if cite == citi:
            cities.remove(cite)
            city_nk = cities
print(city_k)
print(city_nk)
            
```
